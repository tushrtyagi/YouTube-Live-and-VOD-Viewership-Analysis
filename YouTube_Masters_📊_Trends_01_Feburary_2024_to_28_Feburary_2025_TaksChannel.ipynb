## VOD View Count

```python
import pandas as pd
import plotly.express as px

# Ensure Pub_Date is in datetime format
filtered_df['Pub_Date'] = pd.to_datetime(filtered_df['Pub_Date'])

# Convert Month to period format for proper sorting
filtered_df['Month'] = filtered_df['Pub_Date'].dt.to_period('M')

# Remove duplicate entries per channel, date, and video
filtered_df.drop_duplicates(subset=['Channel_title', 'Pub_Date', 'Video_id'], inplace=True)

# Group data properly and sum values, including 'Month' in the groupby
grouped_df = (
    filtered_df.groupby(['Channel_title', 'Month'], as_index=False)
    .agg({'View_count': 'sum', 'Favorite_count': 'sum',
          'Comment_count': 'sum', 'Like_count': 'sum',
          'Dislike_count': 'sum'})
)

# Convert Month to string for plotting
grouped_df['Month'] = grouped_df['Month'].astype(str)

# Compute the total view count per channel across all months
channel_order = (
    grouped_df.groupby('Channel_title')['View_count']
    .sum()
    .sort_values(ascending=False)
    .index.tolist()
)

# Set Channel_title as a categorical variable ordered by total view count
grouped_df['Channel_title'] = pd.Categorical(grouped_df['Channel_title'], categories=channel_order, ordered=True)

# Sort the DataFrame by Month and Channel_title (which now follows channel_order)
grouped_df = grouped_df.sort_values(by=['Month', 'Channel_title']).reset_index(drop=True)

# Create a new column that converts View_count to millions (M) and formats it as a string.
grouped_df['View_count_M'] = (grouped_df['View_count'] / 1e6).round(2).astype(str) + 'M'

# Create a bar plot using Plotly Express with value labels above each bar
fig = px.bar(
    grouped_df,
    x='Month',
    y='View_count',
    color='Channel_title',
    barmode='group',  # Display bars side by side
    title='YouTube | VOD | View Count | 1-Feb-2025 to 15-Feb-2025 | Taks Channels',
    labels={'Month': 'Publish Month', 'View_count': 'Total View Count'},
    category_orders={"Channel_title": channel_order},
    color_discrete_map=channel_colors,
    text='View_count_M'  # Use the formatted view count in millions as text labels
)

# Update traces so that the text appears above the bars
fig.update_traces(texttemplate='%{text}', textposition='outside')

# Customize layout for a white background and proper legend placement
fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(t=80, b=50, l=50, r=50),
    title=dict(
        y=0.95,  # Moves the title slightly down
        x=0.5,
        xanchor='center',
        yanchor='top',
        font=dict(size=20)
    ),
    legend=dict(
        orientation='h',  # Horizontal legend
        yanchor='bottom',
        y=0.99,  # Moves legend higher
        xanchor='center',
        x=0.5,
        font=dict(size=10)
    ),
    xaxis=dict(
        showgrid=False,
        gridcolor='lightgray'
    ),
    yaxis=dict(
        showgrid=False,
        gridcolor='lightgray'
    )
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import os
from tqdm import tqdm

# Path to the directory containing the CSV files
directory_path = "/content/drive/MyDrive/VOD_RAW_BACKUP/Oct2025-Jan2025"

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through all files in the directory
for filename in tqdm(os.listdir(directory_path)):
    if filename.endswith(".csv"):  # Check if the file is a CSV file
        file_path = os.path.join(directory_path, filename)
        try:
            print(file_path)
            # Read the CSV file into a DataFrame, specifying the encoding
            df = pd.read_csv(file_path, encoding='latin-1')  # or 'cp1252', or other encodings to try
            dfs.append(df)  # Append the DataFrame to the list
        except pd.errors.EmptyDataError:
            print(f"Warning: Skipping empty file {filename}")
        except pd.errors.ParserError:
            print(f"Warning: Skipping file with parsing error {filename}")
        except UnicodeDecodeError:
            print(f"Warning: Skipping file with encoding error {filename}, tried with latin-1. Try with 'cp1252' or other encodings if this fails")

# Concatenate all DataFrames in the list into a single DataFrame
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print("Combined DataFrame:")
    print(combined_df.head()) #prints first 5 rows
else:
    print("No CSV files found in the specified directory.")
```

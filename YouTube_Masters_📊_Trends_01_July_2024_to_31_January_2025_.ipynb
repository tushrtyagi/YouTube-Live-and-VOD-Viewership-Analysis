# Importing Libraries

```python
import pandas as pd
import plotly.express as px

```

```python
df = pd.read_csv('/content/drive/MyDrive/final_combined_jul24tojan25_df.csv')
df
```

```python
# Define the channel list with corresponding colors
channel_colors = {
    'News18 India': 'black',  # Black
    'Republic Bharat': 'orange',  # Orange
    'IndiaTV': 'grey',  # Grey
    'Zee News': 'blue',  # Blue
    'Aaj Tak': 'red',  # Red
    'News 24': 'green',  # Green
    'TV9 Bharatvarsh': 'purple',  # Purple
    'TIMES NOW Navbharat': 'pink',  # Pink
    'ABP NEWS': 'brown',  # Brown
    'Good News Today': 'lightcoral',  # Light Red
    'NDTV India': 'lightgreen',  # Light Green
    'News Nation': 'yellow'  # Yellow
}
```

```python
# Define the channel list
channel_titles = [
 'News18 India',
 'Republic Bharat',
 'IndiaTV',
 'Zee News',
 'Aaj Tak',
 'News 24',
 'TV9 Bharatvarsh',
 'TIMES NOW Navbharat',
 'ABP NEWS',
 'Good News Today',
 'NDTV India',
 'News Nation',
]

# channel_titles_eng = [
#  'India Today',
#  'MIRROR NOW',
#  'Firstpost',
#  'WION',
#  'TIMES NOW',
#  'NDTV',
#  'Republic World',
#  'CNN-News18'
# ]
```

# Month On Month

# VOD

```python
# Filter only the required channels & Type = "VOD"
filtered_df = df[
    (df['Channel_title'].isin(channel_titles))
    &
    (df["Type"] == "VOD")
    # |
    # (combined_dff_final["Type"] == "Live then VOD")
    &
    (~(df["Shorts"]=="Shorts"))
].copy()
```

## VOD View Count

```python
# Ensure Pub_Date is in datetime format
filtered_df['Pub_Date'] = pd.to_datetime(filtered_df['Pub_Date'])

# Convert Month to period format for proper sorting
filtered_df['Month'] = filtered_df['Pub_Date'].dt.to_period('M')

# Remove duplicate entries per channel, date, and video
filtered_df.drop_duplicates(subset=['Channel_title', 'Pub_Date', 'Video_id'], inplace=True)

# Group data properly and sum values, include 'Month' in the groupby
grouped_df = (
    filtered_df.groupby(['Channel_title', 'Month'], as_index=False)
    .agg({'View_count': 'sum', 'Favorite_count': 'sum',
          'Comment_count': 'sum', 'Like_count': 'sum',
          'Dislike_count': 'sum'})
)

# Convert Month to string for plotting
grouped_df['Month'] = grouped_df['Month'].astype(str)

# Sort data by Month
grouped_df = grouped_df.sort_values(by='Month').reset_index(drop=True)

# Plot using Plotly Express
fig = px.line(
    grouped_df,
    x='Month',
    y='View_count',
    color='Channel_title',
    title='YouTube | VOD | View Count | July 2024 to January 2025 | Hindi News',
    labels={'Month': 'Publish Month', 'View_count': 'Total View Count'},
    category_orders={"Channel_title": list(channel_colors.keys())},
    color_discrete_map=channel_colors
)

# Customize layout for white background and move legend properly
fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(t=80, b=50, l=50, r=50),  # Add more top margin to avoid title overlap
    title=dict(
        y=0.95,  # Moves the title slightly down
        x=0.5,
        xanchor='center',
        yanchor='top',
        font=dict(size=20)  # Adjust title font size
    ),
    legend=dict(
        orientation='h',  # Horizontal legend
        yanchor='bottom',
        y=.97,  # Increase y to move legend higher
        xanchor='center',
        x=0.5,
        font=dict(size=10)  # Adjust legend font size
    ),
    xaxis=dict(
        showgrid=False, gridcolor='lightgray'
    ),
    yaxis=dict(
        showgrid=False, gridcolor='lightgray'
    )
)

# Show the plot
fig.show()
```
